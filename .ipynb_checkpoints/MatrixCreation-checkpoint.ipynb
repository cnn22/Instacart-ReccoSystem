# Library Making

```python
from itertools import combinations 
import pickle
import pandas as pd

def rSubset(arr, r): 
  
    # return list of all subsets of length r 
    # to deal with duplicate subsets use  
    # set(list(combinations(arr, r))) 
    return list(combinations(arr, r)) 

def insertUniquePairingIntoDict(dic, array):
    for pair in rSubset(array, 2):
        if dic.get(pair) == None:
            dic[pair] = 1
        else:
            dic[pair] = dic[pair] + 1
    return dic
    
def openPickle(name):
    file = open(name, 'rb')
    return pickle.load(file)
```

```python
data = openPickle("productPerUser.pickle")
data.head()
```

```python
matrixDict = {}
iterNumber = 1
for productIDs in data.head().ProductID:
    matrixDict = insertUniquePairingIntoDict(matrixDict,productIDs)
    print("Iteration Number: " + str(iterNumber))
    iterNumber += 1
print(matrixDict)
```
